```python
import requests ##importing necessary packages
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://soc.ucla.edu/faculty' ##storing url to a name
```

```python
webpage = requests.get(url) ##retrieving website data
page_content=webpage.content 
```

```python
soup = BeautifulSoup(page_content, "lxml") ##converting data into a beautiful soup
```

```python
faculty_all=soup.findAll("div",id="faculty-directory-wrapper") ##finding tag with faculty data
```

```python
print(faculty_all)
```

```python

faculty_clean=[] ##creating a list to store clean data
for i in faculty_all: ##retrieving individual faculty data
    faculty={} 
    faculty['name']=i.find("h1").getText(strip=True)  
    faculty['title']=i.find("h2").getText(strip=True)
    faculty['subfields']=i.find(class_="span6 faculty-directory-middle-col").getText(strip=True)
    faculty['email']=i.find(class_="faculty-email").getText(strip=True)
    faculty_clean.append(faculty)
print(faculty_clean)
    
```

```python
df = pd.DataFrame(faculty_clean)  ##converting to pandas data frame
df
```
